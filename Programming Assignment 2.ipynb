```python
# 1. Write a Python program to convert kilometers to miles
km = float(input("Enter km: "))
print("%.1f km is %.3f miles" %(km, km*0.621371))
```

```python
# 2. Write a Python program to convert Celsius to Fahrenheit
C = float(input("Enter temperature in Celsius: "))
print("%.1f Celsius is %.2f Fahrenheit" %(C,(C * 9/5) + 32))
```

```python
# 3. Write a Python program to display calendar
import calendar

year = int(input("Enter the year: "))
month = int(input("Enter the month: "))
print(calendar.month(year, month))
```

```python
# 4. Write a Python program to solve quadratic equation

import cmath

print("Standard quadratic equation is given by ax**2 + bx + c = 0")
a = int(input("Enter coefficient of a: "))
b = int(input("Enter coefficient of b: "))
c = int(input("Enter coefficient of c: "))

d = (b**2) - (4*a*c)

solution1 = (-b-cmath.sqrt(d)) / (2*a)
solution2 = (-b+cmath.sqrt(d)) / (2*a)

print("First solution is:", solution1)
print("Second solution is:", solution2)
```

```python
# 5. Write a Python program to swap two variables without temp variable

a = input("Enter a: ")
b = input("Enter b: ")
print("Before Swapping:",a,b)
b, a = a, b 
print("After Swapping:",a,b)
```
